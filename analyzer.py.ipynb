```python
import requests
from utils import get_user_repos, analyze_languages
from graph import plot_language_chart

username = input("Enter GitHub username: ")
repos = get_user_repos(username)

if repos:
    total_stars = sum(repo['stargazers_count'] for repo in repos)
    total_forks = sum(repo['forks_count'] for repo in repos)

    print(f"\nUser: {username}")
    print(f"Public Repositories: {len(repos)}")
    print(f"Total Stars: {total_stars}")
    print(f"Total Forks: {total_forks}")

    lang_stats = analyze_languages(repos)
    print("\nLanguages Used:")
    for lang, count in lang_stats.items():
        print(f"{lang}: {count}")

    plot_language_chart(lang_stats, username)
else:
    print("User not found or no public repositories.")
```
